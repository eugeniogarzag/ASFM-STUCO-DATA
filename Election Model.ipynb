```python
import csv
import random

#ASFM Student Number 2018
EightMale = 76
EightFemale = 83
NineMale = 81
NineFemale = 75
TenMale = 88
TenFemale = 79
ElevenMale = 80
ElevenFemale = 60

#Wins Track
RedWin = 0
NeonWin = 0
OrangeWin = 0

#Certainties and Probablity Weighted
EightRed =(.529) * 17/(EightMale + EightFemale) 
NineRed = (.25) * 28/(NineMale + NineFemale) 
TenRed = (.286) * 35/(TenMale + TenFemale) 
ElevenRed = (.604) * 48/(ElevenMale + ElevenFemale)

EightNeon = (.176) * 17/(EightMale + EightFemale) 
NineNeon = (.464) * 28/(NineMale + NineFemale) 
TenNeon = (.457) * 35/(TenMale + TenFemale) 
ElevenNeon = (.167) * 48/(ElevenMale + ElevenFemale)

EightOrange = (.294) * 17/(EightMale + EightFemale) 
NineOrange = (.286) * 28/(NineMale + NineFemale) 
TenOrange = (.257) * 35/(TenMale + TenFemale) 
ElevenOrange = (.229) * 48/(ElevenMale + ElevenFemale)

Eight = EightRed + EightNeon + EightOrange
Nine = NineRed + NineNeon + NineOrange
Ten = TenRed + TenNeon + TenOrange
Eleven = ElevenRed + ElevenNeon + ElevenOrange




#Automation in progress
#data = csv.reader(open('STUCO.csv', newline=''), delimiter=' ', quotechar='|')

iterations = 500


for x in range(0,iterations):
    #Vote Count
    RedCount = 0
    NeonCount = 0
    OrangeCount = 0
    for x in range(0, (EightMale + EightFemale)):
        rand = random.random()
        if rand < EightRed / Eight:
            RedCount+= 1
        elif rand < (EightRed+EightNeon)/Eight:
            NeonCount+= 1
        else:
            OrangeCount+= 1

    for x in range(0, (NineMale + NineFemale)):
        rand = random.random()
        if rand < NineRed / Nine:
            RedCount+= 1
        elif rand < (NineRed+NineNeon)/Nine:
            NeonCount+= 1
        else:
            OrangeCount+= 1

    for x in range(0, (TenMale + TenFemale)):
        rand = random.random()
        if rand < TenRed / Ten:
            RedCount+= 1
        elif rand < (TenRed+TenNeon)/Ten:
            NeonCount+= 1
        else:
            OrangeCount+= 1

    for x in range(0, (ElevenMale + ElevenFemale)):
        rand = random.random()
        if rand < ElevenRed / Eleven:
            RedCount+= 1
        elif rand < (ElevenRed+TenNeon)/Eleven:
            NeonCount+= 1
        else:
            OrangeCount+= 1

    if RedCount > NeonCount and RedCount > OrangeCount:
        RedWin += 1
    elif NeonCount > RedCount and NeonCount > OrangeCount:
        NeonWin += 1
    else:
        OrangeWin+=1
    print("Red:" + str(RedCount) + " Neon:" + str(NeonCount) + " Orange:" + str(OrangeCount))
        
#FinalWinProbabilityVariable
RedProb = RedWin/iterations * 100
NeonProb = NeonWin/iterations * 100
OrangeProb = OrangeWin/iterations * 100

print()        
print("Red has a " + str(RedProb) + "% chance of winning")
print("Orange has a " + str(OrangeProb) + "% chance of winning")
print("Neon has a " + str(NeonProb) + "% chance of winning")
```
